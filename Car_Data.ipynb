```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(1)

url = "https://www.cars24.com/buy-used-cars-new-delhi/?itm_source=Cars24Website&itm_medium=sticky_header"
driver.get(url)
time.sleep(1)
```

```python
# function for load page
def load_more():
    for i in range(5):
        driver.execute_script("window.scrollBy(0,10000)")
        
#function for collecting link of each result in form of url in "urls" list
urls=[]
def links():
    match_page=driver.find_elements_by_xpath("//div[@class='col-4']/a")
    for i in match_page:
        urls.append(i.get_attribute('href'))
        
        
#viewing more results on page.
for i in range(0,9):
    links()
    load_more()
    time.sleep(2)
```

```python
#collecting data in lists from fetched product url.
Car_Name =[] #collecting car name along with manufacture year
Mix=[] #collecting fuel type,kilometers,Gear type
State=[] #list for collecting location
Price =[] #list for Price of car
Gear_Type=[]#collecting Gear_Type of car
for i in urls:
    driver.get(i)
    
    q=len(Car_Name)
    print(q,'remaining',q-len(Car_Name))
    
    #collecting car name along with manufacture year
    try:
        name=driver.find_element_by_class_name('_2geSF')
        Car_Name.append(name.text)
    except NoSuchElementException:
        Car_Name.append("NA")
    
    #collecting fuel type,kilometers,Gear type
    try:
        pt=driver.find_element_by_class_name('DB7s9')
        Mix.append(pt.text)
    except NoSuchElementException:
        Mix.append("NA")
     
     #Collectring owner info
#     try:
#         own=driver.find_element_by_class_name('_1gasz')
#         owner.append(own.text)
#     except NoSuchElementException:
#         owner.append("NA")
        
    #collecting model name      
#     try:
#         gr=driver.find_element_by_class_name('_18vo2')
#         Gear_Type.append(gr.text)
#     except NoSuchElementException:
#         Gear_Type.append("NA")
        
    #collecting State    
    try:
        loc=driver.find_element_by_class_name('_2NHUv')
        State.append(loc.text)
    except NoSuchElementException:
        State.append("NA")
        
    #Price of car    
    try:
        pr=driver.find_element_by_class_name('_2yYvS')
        Price.append(pr.text)
    except NoSuchElementException:
        Price.append("NA")
```

```python
Car_Name
```

```python
m=[]
CONDITION =[] #collecting vehicle condition
OWNER = []#collecting owner of car
KMS =[]#collecting kms of car
FUEL_TYPE=[]#collecting fuel_type of car
TRANSMISSION = [] #collecting vehicle condition
INS_VALID = [] #collecting vehicle insurance info

for i in Mix:
    m.append(re.split(r'\n',i))
for j in m:
    
    #fetching vehicle condition
    if len(j)>1:
        CONDITION.append(j[1])
    else:
        CONDITION.append("NA")
    
    #fetching vehicle owner
    if len(j)>1:
        OWNER.append(j[3])
    else:
        OWNER.append("NA")
    
    #fetching vehicle kilometers history
    if len(j)>1:
        KMS.append(j[5])
    else:
        KMS.append("NA")
     
    #fetching vehicle fuel type
    if len(j)>1:
        FUEL_TYPE.append(j[7])
    else:
        FUEL_TYPE.append("NA")
    
    #fetching vehicle transmission
    if len(j)>1:
        if len(j)==19:
            TRANSMISSION.append(j[11])
        else:
            TRANSMISSION.append(j[9])
    else:
        TRANSMISSION.append("NA")  
        
        
    #fetching vehicle insurance validity
    INS_VALID = []
    for j in m:
        if len(j)>1:
            if len(j)==19:
                INS_VALID.append(j[15])
            else:
                INS_VALID.append(j[13])
        else:
            INS_VALID.append("NA")
            

print(len(CONDITION))        
print(len(OWNER))
print(len(KMS))
print(len(FUEL_TYPE))
print(len(TRANSMISSION))
print(len(INS_VALID))
```

```python
#splitting info by using re
yr=[]
for i in Car_Name:
    yr.append(re.split(r'\s',i))  
#collecting year and company name
YEAR=[]
COMPANY=[]

for i in yr:
    
    #fetching year
    if len(i)==1:
        YEAR.append("NA")
    else:
        YEAR.append(i[0])  
        
    #fetching Car Name
    
    if len(i)!=1:  
        if  i[1] == 'MG':
            COMPANY.append(i[1] + " " + i[i.index(i[1])+1])
        else:
            COMPANY.append(i[1])
    else:
        COMPANY.append("NA")
print(len(COMPANY))
print(len(YEAR))
```

```python
#collecting State of car
l=[]
LOCATION=[]
for i in State:
    l.append(re.split(r'\s',i))

for i in l:
    if len(i)>1:
        LOCATION.append(i[1])
    else:
        LOCATION.append("NA")
len(LOCATION)
```

```python
#Fetching model name
#copy list
test = Car_Name
test_1=[]


for i in test:
    test_1.append(re.split(r'\s',i))
    

for i in test_1:
    for j in i:
        if j in YEAR or j in TRANSMISSION:
            i.remove(j)
#         elif :
#             i.remove(j)

for i in test_1:
    for j in i:
         if  j in COMPANY:
                i.remove(j)
                
MODEL=[]
for i in test_1:
        MODEL.append(' '.join(i))
```

```python
#fetching only numbers from owner
ows=[]
OW=[]
for i in OWNER:
    ows.append(re.findall(r'[^a-zA-Z]',i))
    
ows
for i in ows:
    OW.append(''.join(i))
    
len(OW)
```

```python
#fetching insurance validity month and year

INSURANCE_UP_TO = []
IO  = []

for i in INS_VALID:
    IO.append(re.split('\s',i))
    
for i in IO:
    if len(i)>1:
        INSURANCE_UP_TO.append(i[-2]+" "+i[-1])
    else:
        INSURANCE_UP_TO.append("NA")      
len(INSURANCE_UP_TO )  
```
